## Número de variables categóricas y número de valores únicos para cada una

```python
def resumen(data):
    
    resultado = pd.DataFrame()

    resultado['COLS'] = data.columns
    resultado['COUNT'] = data.count().values
    resultado['NULL'] = data.isna().sum().values
    resultado['NUNIQUE'] = df_original.nunique().values
    resultado['FREQ'] = data.mode().iloc[0].values
    resultado['MIN'] = data.describe().T['min'].values
    resultado['MAX'] = data.describe().T['max'].values
    resultado['MEAN'] = data.describe().T['mean'].values 
    resultado['STD'] = data.describe().T['std'].values
    
    return resultado
```

# Características de funcionamiento del receptor (ROC)
La curva ROC produce un gráfico único que resume todas estas diferentes tablas de contingencia. La curva ROC representa la tasa de verdaderos positivos ($TPR=\frac{TP}{TP+FN}$) versus la tasa de falsos positivos ($FPR=\frac{FP}{TN+FP}$). 
- TPR es la proporción de ejemplos positivos que se clasifican correctamente como positivos.
- FPR es el número de falsos positivos en relación con el número total de negativos reales.

```python
x_pred.shape
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

# Calcular y trazar la curva ROC para cada columna
num_columnas = arreglo_y_test.shape[1]

for i in range(num_columnas):
    etiquetas = arreglo_y_test[:, i]
    puntuaciones = arreglo_X_test[:, i]
    
    fpr, tpr, umbrales = roc_curve(etiquetas, puntuaciones)
    roc_auc = auc(fpr, tpr)
    
    plt.plot(fpr, tpr, label='Columna %d (AUC = %0.2f)' % (i+1, roc_auc))

# Visualizar la curva ROC para cada columna
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Clasificación al azar')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC para cada columna')
plt.legend(loc="lower right")
plt.show()
```

```python
def confusionMatrix(predicted, actual, threshold):
    if len(predicted) != len(actual): return -1
    tp = 0.0
    fp = 0.0
    tn = 0.0
    fn = 0.0
    
    for i in range(len(actual)): #actual or predicted
        if predicted[i] > 0.5: #Labels that are 1.0 (positive examples)
            if predicted[i] > threshold:
                tp += 1.0 # Correctly predicted positive
            else:
                fn += 1.0 # Incorrectly predicted negative
        else: # Labels that are 0.0 (negative examples)
            if (predicted[i] < threshold): #Labels that are 1.0 (positive examples)
                tn += 1.0 #Correctly predicted negative
            else:
                fp += 1.0 # Incorrectly predicted positive
    matConf = [tp, fn, fp, tn]
    return matConf
```

# Demostración de LassoCV con normalización de datos y sin normalización.

```python
import pandas as pd
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_atributos, df_etiquetas, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import MultiTaskLassoCV

# Sin normalización
lasso_sin_norm = MultiTaskLassoCV(cv=5)
lasso_sin_norm.fit(X_train, y_train)
pred_sin_norm = lasso_sin_norm.predict(X_test)

# Con normalización
scaler = StandardScaler()
X_train_con_norm = scaler.fit_transform(X_train)
X_test_con_norm = scaler.transform(X_test)
lasso_con_norm = MultiTaskLassoCV(cv=5)
lasso_con_norm.fit(X_train_con_norm, y_train)
pred_con_norm = lasso_con_norm.predict(X_test_con_norm)
```

```python
print("Coeficientes sin normalización:", lasso_sin_norm.coef_.shape)
print("Coeficientes con normalización:", lasso_con_norm.coef_.shape)
print("Valores de alpha utilizados sin normalización:", lasso_sin_norm.alphas_.shape)
print("Valores de alpha utilizados con normalización:", lasso_con_norm.alphas_.shape)
```

```python
import matplotlib.pyplot as plt

# Coeficientes sin normalización
coef_sin_norm = lasso_sin_norm.coef_
plt.figure(figsize=(10, 6))
plt.plot(range(len(coef_sin_norm)), coef_sin_norm)
plt.xlabel('Índice de característica')
plt.ylabel('Coeficiente')
plt.title('Coeficientes de LassoCV sin normalización')
plt.show()

# Coeficientes con normalización
coef_con_norm = lasso_con_norm.coef_
plt.figure(figsize=(10, 6))
plt.plot(range(len(coef_con_norm)), coef_con_norm)
plt.xlabel('Índice de característica')
plt.ylabel('Coeficiente')
plt.title('Coeficientes de LassoCV con normalización')
plt.show()
```

```python
from sklearn.linear_model import Lasso

# Generar secuencia de alfas para la ruta de regularización
alphas_sin_norm = lasso_sin_norm.alphas_
coefs_sin_norm = []
for a in alphas_sin_norm:
    lasso = Lasso(alpha=a)
    lasso.fit(X_train, y_train)
    coefs_sin_norm.append(lasso.coef_)
coefs_sin_norm = np.array(coefs_sin_norm)

# Graficar ruta de regularización sin normalización
plt.figure(figsize=(10, 6))
for i in range(coefs_sin_norm.shape[1]):
    plt.plot(-np.log10(alphas_sin_norm), coefs_sin_norm[:, i])
plt.xlabel('-log10(alfa)')
plt.ylabel('Coeficiente')
plt.title('Ruta de regularización de LassoCV sin normalización')
plt.axis('tight')
plt.show()

# Generar secuencia de alfas para la ruta de regularización con normalización
alphas_con_norm = lasso_con_norm.alphas_
coefs_con_norm = []
for a in alphas_con_norm:
    lasso = Lasso(alpha=a)
    lasso.fit(X_train_con_norm, y_train)
    coefs_con_norm.append(lasso.coef_)
coefs_con_norm = np.array(coefs_con_norm)

# Graficar ruta de regularización con normalización
plt.figure(figsize=(10, 6))
for i in range(coefs_con_norm.shape[1]):
    plt.plot(-np.log10(alphas_con_norm), coefs_con_norm[:, i])
plt.xlabel('-log10(alfa)')
plt.ylabel('Coeficiente')
plt.title('Ruta de regularización de LassoCV con normalización')
plt.axis('tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

# Concatenar los conjuntos de entrenamiento y prueba
X = pd.concat([X_train, X_test])
y = pd.concat([y_train, y_test])

# Opcional: Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Crear una instancia de Lasso
lasso = Lasso()

# Generar secuencia de alfas para la ruta de regularización
alphas = np.logspace(-4, 0, 100)

# Ajustar el modelo Lasso para cada valor de alfa en la secuencia
coefs = []
for a in alphas:
    lasso.set_params(alpha=a)
    lasso.fit(X_scaled, y)
    coefs.append(lasso.coef_)
coefs = np.array(coefs)

# Graficar la ruta de regularización
plt.figure(figsize=(10, 6))
for i in range(coefs.shape[1]):
    plt.plot(-np.log10(alphas), coefs[:, i])
plt.xlabel('-log10(alfa)')
plt.ylabel('Coeficiente')
plt.title('Ruta de regularización de Lasso')
plt.axis('tight')
plt.show()
```
